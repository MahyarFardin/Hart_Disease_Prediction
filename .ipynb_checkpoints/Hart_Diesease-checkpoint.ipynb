# ML


```python
x_scaled_regressive = pd.DataFrame(mms.fit_transform(x[x.columns[not_check_list]]), index = df.index)
final_x_train = pd.concat([x[x.columns[check_list]], x_scaled_regressive], axis = 1)
```

```python
def full_model(x, y, model, params):
  shuffle = ShuffleSplit(n_splits=10, test_size=.1)
  rgs = RandomizedSearchCV(model, params, cv=shuffle, n_jobs = -1)
  rgs.fit(x, y)
  print("Estimator \n", rgs.best_estimator_)
  print("Best score \n", rgs.best_score_)
```

```python
rfc = RandomForestClassifier()
parameters_rfc = {"n_estimators" : [400],
                  "criterion" : ["gini", "entropy"],
                  "max_depth" : [2, 4, 6, 8, 10],
                  "min_samples_split" : [2, 4, 8],
                  "min_samples_leaf" : [2, 4, 6, 8]}
full_model(x_scaled_regressive, y, rfc, parameters_rfc)
```

```python
adb = AdaBoostClassifier()
lr = LogisticRegression()
parameters_adb = {"estimator" : [lr, ],
                  "n_estimators" : [300, 400, 500]}
full_model(x_scaled_regressive, y, adb, parameters_adb)
```

```python
svm = SVC()
parameters_svm = {"C" : [.1, .5, 1.0, 1.5, 2, 5, 10],
                  "kernel" : ["linear", "poly", "rbf", "sigmoid"],
                  "degree" : [1, 2, 3, 4, 5, 6, 7] ,
                  "gamma" : ['scale', "auto"],
                  "shrinking" : [True, False],
                  "probability" : [True, False],
                  "tol" : [10, 1, .1, .01, 0.001, .0001]}
full_model(x_scaled_regressive, y, svm, parameters_svm)
```

```python
bn = BernoulliNB()
vc = AdaBoostClassifier(base_estimator = bn, n_estimators = 400, learning_rate=1)
vc.fit(final_x_train, ytrain)
```

```python
vc_prediction = vc.predict(final_x_test)
```
